```python
s1, s2, s3 = input("Masukkan string = ").split()
a, b, c = map(len, (s1, s2, s3))

if a < b > c:
    print("Yeehaaw")
elif (a < b < c) or (a > b > c):
    print("Woohoohoo")
elif a == b or b == c or c == a:
    print("Hebwueh")
```

```python
skor = int(input("Masukkan skor = "))

if skor < 899 :
  print("Bronze")
elif skor >= 899 and skor <= 1097 :
  print("Warrior")
elif skor >= 1098 and skor <= 3500 :
  print("Mythic")
elif skor >= 3501 and skor <= 4737 :
  print("Legend")
elif skor > 4737:
  print("Epic")
```

```python
S = input("Masukkan string = ").strip()

alphabet = "abcdefghijklmnopqrstuvwxyz"

hasil_kali = 1

for char in S :
  if char.isalpha():
    lowercase = char.lower()
    nilai = alphabet.index(lowercase) + 1
    hasil_kali *= nilai
  elif char.isdigit():
    hasil_kali *= int(char)

print(hasil_kali)
```

```python
N = int(input())

if N % 2 == 0:
  N_baru = (1/2*N)
  print(N_baru)
elif N % 2 == 1:
  N_baru = (3/4*N)
  print(N_baru)
```

```python
N = int(input())
M = int(input())

for i in range(N):
  for j in range(M):
    if (i + j) % 2 == 0:
      print('**', end='')
    else :
      print('##', end='')
  print()
```

```python
asli = input().strip()
berubah = input().strip()

alphabet = 'abcdefghijklmnopqrstuvwxyz'

for i in range(1,26):
  geser = ""
  for char in asli:
    if char.isalpha():
      huruf = alphabet.index(char.lower())
      new_char = alphabet[(huruf + i) % 26]
      if char.isupper():
        geser += new_char.upper()
      else:
        geser += new_char
    else:
      geser += char

  sandi = geser[::-1]

  if sandi == berubah:
    print(f"Silahkan berjalan sejauh {i} meter")
    break

if sandi != berubah:
  print(f"Mending pulang aja")
```
