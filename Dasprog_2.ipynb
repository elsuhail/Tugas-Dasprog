```python
N = int(input())

for i in range(N):
  jumlah_baris = int(input())
  segitiga = []
  for j in range(jumlah_baris):
    baris = list(map(int, input().split()))
    segitiga.append(baris)

    if len(segitiga[j]) != j+1:
      break
  for k in range(jumlah_baris - 2, -1, -1):
    for m in range(len(segitiga[k])):
      segitiga[k][m] += max(segitiga[k+1][m], segitiga[k+1][m+1])

    print(segitiga[0][0])
```

```python
x = int(input())

perfect_fences_sebelumnya = 0
perfect_fences_sekarang = None

for i in range(1,33):
  val = 2**i-1
  if val >= x:
    perfect_fences_sekarang = val
    break
  perfect_fences_sebelumnya = val

selisih_sebelumnya = abs(x - perfect_fences_sebelumnya)
selisih_sekarang = abs(x - perfect_fences_sekarang)

if x == perfect_fences_sekarang or x == perfect_fences_sebelumnya:
  print("It's the perfect amount, nice!")
elif x > perfect_fences_sebelumnya and x < perfect_fences_sekarang:
  if selisih_sebelumnya < selisih_sekarang:
    print(f"Maybe it's better to remove {selisih_sebelumnya} fences, just to be perfect")
  else:
    print(f"Maybe it's better to add {selisih_sekarang} fences, just to be perfect")
```
